1. Cari bagaimana cara menampilkan Judul (title) dari grafik\
   Jawab: dengan menggunakan perintah
          plt.title("TITLE")
   

```python
#Contoh
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([10,20,30,40,50,60,70,80])

plt.title("PEMASUKAN TAHUN 2021") #jika ingin bervariasi bisa menambah perintah dengan menambahkan koma(,) di belakang (")

plt.plot(x,y)
plt.show()
```

2. Cari bagaimana cara memberikan label pada sumbu x dan sumbu y\
   Jawab: dengan menggunakan perintah
          plt.xlabel("LABEL")
          plt.ylabel("LABEL")

```python
#Contoh
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([10,20,30,40,50,60,70,80])

plt.title("PEMASUKAN TAHUN 2021")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Pemasukan")

plt.plot(x,y)
plt.show()
```

3. Cari bagaimana cara menambahkan teks dalam grafik\
   Jawab: dengan menggunakan perintah 
          plt.text("TEXT")

```python
#Contoh
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([40,20,10,40,70,60,10,80])

plt.title("PEMASUKAN TAHUN 2021")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Pemasukan")
plt.text(1,40,"40")

plt.plot(x,y)
plt.show()
```
